### PLOTTING & MAPPING ###

Now I am going to upload the 2020 Census Tract geojson file for San Francisco. This will allow me to map the COVID-19 data.

```python
import folium
```

```python
# read the SF Census Tracts geojson file

tracts=gpd.read_file('week04 data/Census 2020_ Tracts for San Francisco (1).geojson')
```

```python
tracts.head()
```

```python
# plot it!
tracts.plot(figsize=(12,10))
```

```python
# I only need the FIPS and geometry information. In this set geoid is the same as the FIPS code.
tracts = tracts[['geoid','geometry']]
```

```python
# renaming the columns so that I can merge the COVID-19 data and census tract file using the FIPS column
tracts.columns = ['FIPS','geometry']
```

```python
# making sure that the columns were sucessfully renamed
tracts.head()
```

```python
#renaming the COVID-19 data file column from id to FIPS
df = df.rename(columns={'id': 'FIPS'})
```

```python
df.head()
```

```python
# create a new dataframe based on the join
tracts_COVID=tracts.merge(df,on="FIPS")
```
